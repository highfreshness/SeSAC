```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('kakao_영등포구맛집(카카오api).csv')
df.head()
```

```python
df.drop(['Unnamed: 0', 'Unnamed: 0.1'], inplace=True, axis=1)
```

```python
df.head()
```

```python
headers = {'UserAgent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.5112.87 Whale/3.16.138.22 Safari/537.36'}
```

```python
response = requests.get('https://place.map.kakao.com/20908598', headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
for x in df['place_url']:
    response = requests.get(x, headers=headers)
    soup = BeautifulSoup(response, 'html.parser')
    print(soup.prettify())
```
