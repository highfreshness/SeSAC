```python
import pandas as pd

pew = pd.read_csv('../datascience/data/pew.csv')
print(pew.head())
print(pew.shape)
```

* melt 메서드 인자
  * id_vars = 위치를 그대로 유지할 열의 이름을 지정
  * value_vars = 행으로 위치를 변경할 열의 이름을 지정
  * var_name = value_var로 위치를 변경한 열의 이름을 지정
  * value_name = var_name 으로 위치를 변경한 열의 데이터를 저장한 열의 이름을 저장

```python
print(pew.iloc[:,0:6])
```

```python
pew_long = pd.melt(pew, id_vars='religion')
print(pew_long.tail())
print(pew_long.shape)
```

```python
pew_long = pd.melt(pew, id_vars='religion', var_name='income', value_name='count')
print(pew_long.head())
print(pew_long.shape)
```

```python
pew_long.to_csv('pew_long.csv', encoding='utf-8-sig')
```

```python
ebola = pd.read_csv('../datascience/data/country_timeseries.csv')
print(ebola.columns)
print(ebola.iloc[:5, [0,1,2,3,10,11]])
```

```python
ebola_long = pd.melt(ebola, id_vars=['Date', 'Day'])
print(ebola_long)
```

```python
variable_split = ebola_long['variable'].str.split('_')
print(variable_split[:5])
```

```python
print(type(variable_split))
print(type(variable_split[0]))
```

```python
status_values = variable_split.str.get(0)
country_values = variable_split.str.get(1)
```

```python
print(status_values[:5])
print(status_values[-5:])
```

```python
print(country_values[-5:])
print(country_values[:5])
```

```python
ebola_long['status'] = status_values
ebola_long['country'] = country_values
print(ebola_long.head())
```

```python
# concat 메서드로 한번에 추가하기
variable_split = ebola_long.variable.str.split('_', expand=True)
variable_split.columns = ['status', 'country']
ebola_parsed = pd.concat([ebola_long, variable_split], axis=1)
```

```python
weather = pd.read_csv('../datascience/data/weather.csv')
print(weather.iloc[:5,:])
print(weather.shape)
```

```python
weather_melt = pd.melt(weather, id_vars=['id', 'year', 'month', 'element'], var_name='day', value_name='temp')
print(weather_melt.tail())
print(weather.shape)
```

```python
weather_tidy = weather_melt.pivot_table(index=['id','year','month','day'], columns='element', values='temp')
weather_tidy
```

```python
billboard = pd.read_csv('../datascience/data/billboard.csv')
billboard_long = pd.melt(billboard, id_vars=['year','artist','track','time','date.entered'], var_name='week', value_name='rating')
print(billboard_long.shape)
```

```python
billboard_long[billboard_long['track']=='Loser'].head()
```

```python
billboard_songs = billboard_long[['year','artist','track','time']]
billboard_songs.shape
```

```python
billboard_songs = billboard_songs.drop_duplicates()
billboard_songs.shape
```

```python
billboard_songs['id'] = range(len(billboard_songs))
billboard_songs.head()
```

```python
billboard_ratings = billboard_long.merge(billboard_songs, on=['year','artist','track', 'time'])
print(billboard_ratings.head())
billboard_ratings.shape
```
