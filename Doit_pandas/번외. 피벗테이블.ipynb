```python
import pandas as pd
```

```python
prac = pd.read_csv('../datascience/data/practice.csv', encoding='utf-8', index_col=0)
prac
```

```python
prac.loc[6] = ['19-3-4','걷기', 1.5]
prac.loc[7] = ['19-3-5','달리기', 1.]
prac.loc[8] = ['19-3-6','걷기', 2.5]
prac.loc[9] = ['19-3-6','달리기', 0.5]
prac
```

```python
# 피벗테이블은 기본 값은 평균을 구하게 되고 다른 집계함수를 선택하려면 aggfunc를 인자를 조절하면 된다
# len = 양(count), 길이
prac.pivot_table(index='운동', aggfunc=[sum, len])
```

```python
# 칼럼 쉽게 지정
prac_pivot = prac.pivot('날짜','운동','양')
prac_pivot.fillna(0, inplace=True)
prac_pivot
```
