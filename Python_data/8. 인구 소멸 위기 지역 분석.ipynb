```python
population = pd.read_excel('DS/data/05. population_raw_data.xlsx', header=1)
population.head()
```

```python
# NaN 값이 NaN값 위의 데이터를 따라 생성됨
population.fillna(method='pad', inplace=True)
population.head()
```

```python
population.rename(columns={'행정구역(동읍면)별(1)':'광역시도', '행정구역(동읍면)별(2)':'시도', '계':'인구수'}, inplace=True)

population = population[(population['시도'] != '소계')] # 소계 행(raw)을 삭제
population
```

```python
population.rename(columns= {'항목':'구분'}, inplace=True)

population.loc[population['구분'] == '총인구수 (명)', '구분'] = '합계'
population.loc[population['구분'] == '남자인구수 (명)', '구분'] = '남자'
population.loc[population['구분'] == '여자인구수 (명)', '구분'] = '여자'

population.head()
```

```python
# 구분 기준은 가임기 여성
population['20-39세'] = population['20 - 24세'] + population['25 - 29세'] + population['30 - 34세'] + population['35 - 39세']
population['65세이상'] = population['65 - 69세'] + population['70 - 74세'] + population['75 - 79세'] + population['80 - 84세'] \
                        + population['85 - 89세'] + population['90 - 94세'] + population['95 - 99세'] + population['100+']
                        
population.head()
```

```python
pop  = pd.pivot_table(population, index=['광역시도','시도'], columns=['구분'], values = ['인구수','20-39세','65세이상'])
pop
```

```python
# pop['소멸비율'] = pop['20-39세','여자'] / (pop['65세이상','여자'] / 2)
# pop.head()
pop['소멸비율'] = pop['20-39세','여자'] / (pop['65세이상','합계'] / 2)
pop.head()
```

```python
pop['소멸위기지역'] = pop['소멸비율'] < 1.0
pop.head()
```

```python
si_name = [None] * len(pop)

tmp_gu_dict = {
    '수원':['장안구','권선구','팔달구','영통구'],
    '성남':['수정구','중원구','분당구'],
    '안양':['만안구','동안구'],
    '안산':['상록구','단원구'],
    '고양':['덕양구','일산동구','일산서구'],
    '용인':['처인구','기흥구','수지구'],
    '청주':['상당구','서원구','흥덕구','청원구'],
    '천안':['동남구','서북구'],
    '전주':['완산구','덕진구'],
    '포항':['남구','북구'],
    '창원':['의창구','성산구','진해구','마산합포구','마산회원구'],
    '부천':['오정구','원미구','소사구']    
}
```

```python
# 특이케이스를 제외하면서 지역이름을 지도 json 파일에 맞춰 정제
for n in pop.index:
    if pop['광역시도'][n][-3:] not in ['광역시', '특별시',' 자치시'] : # 광역시도 이름의 뒤에 광역/특별/자치시가 들어가지 않은 경우
        if pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='강원도': # 강원도 고성을 고성(강원)으로 변경
            si_name[n] = '고성(강원)'
        elif pop['시도'][n][:-1]=='고성' and pop['광역시도'][n]=='경상남도': # 경상남도 고성을 고성(경남)으로 변경
            si_name[n] = '고성(경남)'
        else:
            si_name[n] = pop['시도'][n][:-1] # 나머지는 군/구의 끝 1글자를 제외하고 si_name[n]에 입력
            
        for keys, values in tmp_gu_dict.items():
            if pop['시도'][n] in values: # tmp_gu_dict value 안에 지역이름이 있는 경우 아래 조건 실행
                if len(pop['시도'][n])==2: # 시의 이름 길이가 2인 경우
                    si_name[n] = keys + ' ' + pop['시도'][n] # 시이름 + 공백 + 구/군 
                elif pop['시도'][n] in ['마산합포구','마산회원구']: # 시의 이름이 마산합포구, 마산회원구에 속할 경우 마산을 제외
                    si_name[n] = keys + ' ' + pop['시도'][n][2:-1] # 시이름 + 공백 + 마산을 제외한 구/군 
                else:
                    si_name[n] = keys + ' ' + pop['시도'][n][:-1] # 구/군 끝 한자리 제외
    
    elif pop['광역시도'][n] == '세종특별자치시': # 광역/특별/자치 시 중 세종시의 이름은 세종으로 변경
        si_name[n] = '세종'
        
    else :
        if len(pop['시도'][n]) == 2: # 광역/자치/특별시도 아니고 세종시도 아닌 경우 
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n] 
        else :
            si_name[n] = pop['광역시도'][n][:2] + ' ' + pop['시도'][n][:-1]
            
si_name
```

```python
pop['ID'] = si_name
```

```python
pop.drop('20-39세남자', axis=1, inplace=True)
pop.drop('65세이상남자', axis=1, inplace=True)
pop.drop('65세이상여자', axis=1, inplace=True)

pop.head()
```
