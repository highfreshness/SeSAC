## 이제는 tqdm_notebook 대신 tqdm으로 통합해 사용

```python
from tqdm import tqdm
import time 

rank = []
main_menu = []
cafe_name = []
url_add = []

list_soup = soup.find_all('div', 'sammy')
bar_total = tqdm(list_soup)

for x in bar_total:
    rank.append(x.find(class_='sammyRank').get_text()) # 순위
    main_menu.append(x.find(class_='sammyListing').get_text().split('\n')[0]) # 메뉴 이름
    cafe_name.append(x.find(class_='sammyListing').get_text().split('\n')[1]) # 가게 이름
    # url_add.append(url_base + x.find('a')['href']) # url
    url_add.append(urljoin(url_base, x.find('a')['href'])) # url
    time.sleep(0.5)
```

## 다수의 페이지에 접근해서 원하는 정보 가져오기

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request # Request = 요청, urlopen = url 열어서
```

```python
df = pd.read_csv('DS/data/03. best_sandwiches_list_chicago(220725).csv', index_col=0)
df.head()
```

```python
# header 없이 스크래핑 시도할 경우 404 forbidden 에러 나타남
url_req = df['URL'][0]
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 \
            (KHTML, like Gecko) Chrome/102.0.5005.141 Whale/3.15.136.29 Safari/537.36'}
req = Request(url_req, headers=headers)
page_1 = urlopen(req)

soup_tmp = BeautifulSoup(page_1, 'html.parser')
soup_tmp
```

```python
print(soup_tmp.find('p', 'addy')) # 가격, 주소, 전화, 사이트 URL
```

```python
price = []
address = []
cafe_url = []

for n in tqdm(df.index[:3]):
    url = Request(df['URL'][n], headers=headers)
    html = urlopen(url)
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    x = soup_tmp.find('p', 'addy').get_text().split()
    price.append(x[0][:-1])
    address.append(' '.join(x[1:-2]))
    cafe_url.append(x[-1])
    
print(len(price), len(address), len(cafe_url))
```

```python
price = []
address = []
cafe_url = []

for n in tqdm(df.index):
    url = Request(df['URL'][n], headers=headers)
    html = urlopen(url)
    soup_tmp = BeautifulSoup(html, 'html.parser')
    
    x = soup_tmp.find('p', 'addy').get_text().split()
    price.append(x[0][:-1])
    address.append(' '.join(x[1:-2]))
    cafe_url.append(x[-1])
    
print(len(price), len(address), len(cafe_url))
```

```python
df.head()
```

```python
df['Price'] = price
df['Address'] = address
df['Cafe_url'] = cafe_url

df = df.loc[:, ['Rank', 'Cafe', 'Menu', 'Price', 'Address', 'Cafe_url']]
df.set_index('Rank', inplace=True)
df.head()
```

```python
df.to_csv('DS/data/03. best_sandwiches_list_chicago2(220725).csv', sep=',', encoding='UTF-8-SIG')
```

## 맛집 위치 지도에 표기하기

```python
import folium
from folium.plugins import MarkerCluster
import googlemaps
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DS/data/03. best_sandwiches_list_chicago2(220725).csv', index_col=0)
df.head()
```

```python
gmaps_key = '**'
gmaps = googlemaps.Client(key=gmaps_key)
```

```python
# Test code
target = '2109 W. Chicago Ave., Cicago'
gmaps_output = gmaps.geocode(target)
gmaps_output
```

```python
gmaps_output[0].get('geometry')['location']['lat']
```

```python
from tqdm import tqdm
```

```python
lat = []
lng = []

for n in tqdm(df.index):
    if df['Address'][n] != 'Multiple':
        target_name = df['Address'][n] + 'Cicago'
        gmaps_output = gmaps.geocode(target_name)
        location_output = gmaps_output[0].get('geometry')
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
        
    else:
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
df['lat'] = lat
df['lng'] = lng

df.head()
```

```python
df[df['lat'].isna()]
```

```python
# Why?... 왜 위도 경도의 평균을 구해 마커를 찍는 것 인지?
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)

folium.Marker([df['lat'].mean(), df['lng'].mean()], popup='center').add_to(mapping)

mapping
```

```python
for x in df.index:
    if df['Address'][x] != 'Multiple':
        folium.Marker([df['lat'][x], df['lng'][x]], popup=df['Cafe'][x]).add_to(mapping)
        
mapping
```

```python
mapping = folium.Map(location=[df['lat'].mean(), df['lng'].mean()], zoom_start=11)
mc = MarkerCluster(control=False) # Coltrol = Whether the Layer will be included in LayerControls.

mapping.add_child(mc)

for x in df.index:
    if df['Address'][x] != 'Multiple':
        folium.Marker([df['lat'][x], df['lng'][x]], popup=df['Cafe'][x]).add_to(mc)

mapping
```
