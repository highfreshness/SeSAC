```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show
```

```python
# y = ax + b
# 인구수와 소계로 a(기울기)와 b(y절편)를 구함
fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
fp1
```

```python
# polynomial == 다항식
# 매개변수로 부터 모델 생성
f1 = np.poly1d(fp1)
# x를 생성(시작, 끝, 갯수)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
#
plt.plot(fx, f1(fx), ls='dashed', lw=2, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show
```

```python
# 선의 기울기와 y절편을 가진 f1에 인구수를 넣고 소계와 차이를 구하면 해당 구가 기준점에서 얼마나 멀어졌는지 구할 수 있다.
data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(15,15))
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

# plt.text(x, y, text, fontsize)
# 실제 scatter 좌표 기준으로 글이 써질 좌표를 구한 뒤 텍스트를 입력
for x in range(10):
    plt.text(df_sort['인구수'][x] * 1.02, df_sort['소계'][x] * 0.98, df_sort.index[x], fontsize=20)
    
plt.xlabel('인구수')
plt.ylabel('인구당비율')

plt.colorbar() # 우측 컬러바
plt.grid() # 격자
plt.show()
```
