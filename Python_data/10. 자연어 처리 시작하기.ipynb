# Naive Bayes Classifier의 이해 - 한글

```python
from konlpy.tag import Okt
```

```python
pos_tagger = Okt()
```

```python
train = [
    ('메리가 좋아', 'pos'), 
    ('고양이도 좋아', 'pos'),
    ('난 수업이 지루해', 'neg'),
    ('메리는 이쁜 고양이야', 'pos'),
    ('난 마치고 메리랑 놀거야', 'pos')
]
```

```python
all_words = set(word.lower() for sentence in train for word in word_tokenize(sentence[0]))
all_words
```

```python
t = [({word : (word in word_tokenize(x[0])) for word in all_words}, x[1]) for x in train]
t
```

```python
classifier = nltk.NaiveBayesClassifier.train(t)
classifier.show_most_informative_features()
```

```python
test_sentesce2 = '난 수업이 마치면 메리랑 놀거야'
```

```python
test_sent_features2 = { word.lower() : (word in word_tokenize(test_sentesce2.lower())) for word in all_words}
test_sent_features2
```

```python
classifier.classify(test_sent_features2)
```

```python
# norm = 정규화(Normalize), stem = 원형
def tokenize(doc):
    return ['/'.join(t) for t in pos_tagger.pos(doc, norm=True, stem=True)] # Okt.pos_tagger
```

```python
print(pos_tagger.pos('이것도 재미있습니당ㅋㅋ'))
print('-------------------------------------------------------------')
print(pos_tagger.pos('이것도 재미있습니당ㅋㅋ', norm=True))
print('-------------------------------------------------------------')
print(pos_tagger.pos('이것도 재미있습니당ㅋㅋ', norm=True, stem=True))
```

```python
train_docs = [(tokenize(row[0]), row[1]) for row in train]
train_docs
```

```python
tokens = [t for d in train_docs for t in d[0]]
tokens
```

```python
def term_exists(doc):
    return {word : (word in set(doc)) for word in tokens}
```

```python
# '메리가 좋아', '고양이도 좋아'
# 토크나이징한 후 문장이 말뭉치에 있는지 확인
train_xy = [(term_exists(d), c) for d,c in train_docs]
train_xy
```

```python
# 테스트
for d in train_docs:
    print(d[0])
    for t in d[0]:
        print(t)
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_xy)
```

```python
test_sentence = [('난 수업이 마치면 메리랑 놀거야')]
```

```python
test_docs = pos_tagger.pos(test_sentence[0])
test_docs
```

```python
test_sent_features3 = {word : (word in tokens) for word in test_docs}
test_sent_features3
```

# 문장의 유사도 측정

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# 단어장에 포함되기 위한 최소 빈도, 디폴트 1
# 1번 이상 들어간 단어만 포함
vectorizer = CountVectorizer(min_df=1)
```

```python
contents = ['메리랑 놀러가고 싶지만 바쁜데 어떻하죠?',
                '메리는 공원에서 산책하고 노는 것을 싫어해요',
                '메리는 공원에서 노는 것도 싫어해요. 이상해요.',
                '먼 곳으로 여행을 떠나고 싶은데 너무 바빠서 그러질 못하고 있어요']
```

```python
X = vectorizer.fit_transform(contents)
vectorizer.get_feature_names()
```

```python
# x안의 feature들을 one-hot벡터 형태로 변환 후 축 변경
X.toarray().transpose()
```

```python
# 문자를 숫자 형태로 변환(vectorize)
print(len(X.toarray().transpose()))
print(len(vectorizer.get_feature_names()))
```

```python
from konlpy.tag import Okt
okt = Okt()
```

```python
contents_tokens = [okt.morphs(row) for row in contents]
```

```python
contents_for_vectorize = []

for content in contents_tokens:
    sentence = ''
    for word in content:
        sentence = sentence + ' ' + word
        
    contents_for_vectorize.append(sentence)
    
contents_for_vectorize
```

```python
X = vectorizer.fit_transform(contents_for_vectorize)
num_samples, num_features = X.shape
num_samples, num_features
```

```python
print(X.shape)
```

```python
new_post = ['메리랑 공원에서 산책하고 놀고 싶어요']
new_post_vec = vectorizer.transform(new_post)
new_post_vec.toarray()
```

```python
# linalg.norm() 명령으로 벡터의 길이를 계산
import scipy as sp

def dist_raw(v1,v2):
    delta = v1 - v2
    return sp.linalg.norm(delta.toarray())
```

```python
# new_post_vec
best_doc = None
best_dist = 65535
best_i = None

for i in range(0, num_samples):
    post_vec = X.getrow(i)
    d = dist_raw(post_vec, new_post_vec)
    print(f'== Post {i} with dist={d:.2f}     : {contents[i]}')
    if d < best_dist:
        best_dist = d
        best_i = i
```

```python
print(f'Best post is {best_i}, dist = {best_dist}')
print('--> new_post', new_post)
print('---->', contents[best_i])
```
