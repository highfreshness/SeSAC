```python
# PIL ( Python Imaging Library )
from wordcloud import WordCloud, STOPWORDS

import numpy as np
from PIL import Image
```

```python
text = open('DS/data/09. alice.txt').read()
alice_mask = np.array(Image.open('DS/data/09. alice_mask.png'))

stopwords = set(STOPWORDS)
stopwords.add('said')
```

```python
plt.figure(figsize=(8,8))
plt.imshow(alice_mask, cmap=plt.cm.gray) # , interpolation='bilinear'
plt.axis('off') # x, y 그래프 표기 안함
plt.show()
```

```python
# 배경 설정
wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)
# 텍스트 추출
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
text = open('DS/data/09. a_new_hope.txt').read()

text = text.replace('HAN', 'Han')
text = text.replace("LUKE'S", 'Luke')

mask = np.array(Image.open('DS/data/09. stormtrooper_mask.png'))
```

```python
# 불용어 중복 제거 및 추가
stopwords = set(STOPWORDS)
stopwords.add('int')
stopwords.add('ext')
```

```python
# 배경 설정
wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, margin=10, random_state=1).generate(text)
default_colors = wc.to_array()
```

```python
# hsi : 색상(hue), 채도(saturation), 명도(Lightness)
# http://jun.hansung.ac.kr/CWP/htmls/html%20hsl%20colors.html
import random
def grey_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    return 'hsl(0, 0%%, %d%%)' % random.randint(60,100)
```

```python
plt.figure(figsize=(12,12))
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3), interpolation='bilinear')
plt.axis('off')
plt.show() 
```
