```python
# 2017년 5월 1일 부터 100개를 생성 freq='D'는 일자 기준 (= 2017-5-1부터 100일 생성)
date = pd.date_range('2017-5-1', periods=100, freq='D')
date
```

```python
from tqdm import tqdm
import urllib

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date):
    html = 'https://movie.naver.com' + '/movie/sdb/rank/rmovie.naver?sel=cur&date={date}'
    response = urlopen(html.format(date=urllib.parse.quote(today.strftime('%Y%m%d'))))
    soup = BeautifulSoup(response, 'html.parser')
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for n in range(0, end)])
    movie_name.extend([soup.find_all('div','tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td','point')[n].string for n in range(0, end)])    
```

append는 리스트를 추가할 경우 리스트 요소 자체를 추가해서 리스트 속의 리스트를 만들지만 
extend는 리스트를 추가해도 기존 리스트 요소에 이어서 추가할 수 있다.

x = [1,2,3]
x.append([4,5]) == [1,2,3, [4,5]]

x = [1,2,3]
x.extend([4,5]) == [1,2,3,4,5]

```python
from tqdm import tqdm
import urllib

movie_date = []
movie_name = []
movie_point = []

for today in tqdm(date):
    html = 'https://movie.naver.com' + '/movie/sdb/rank/rmovie.naver?sel=cur&date=' + today.strftime('%Y%m%d')
    response = urlopen(html)
    soup = BeautifulSoup(response, 'html.parser')
    end = len(soup.find_all('td', 'point'))
    
    movie_date.extend([today for _ in range(0, end)])
    movie_name.extend([soup.find_all('div','tit5')[n].a.string for n in range(0, end)])
    movie_point.extend([soup.find_all('td','point')[n].string for n in range(0, end)])    
```

```python
print(len(movie_date), len(movie_name), len(movie_point))
```

```python
movie = pd.DataFrame({'date':movie_date, 'name':movie_name, 'point':movie_point})
movie.head()
```

```python
movie.info()
```

```python
movie['point'] = movie['point'].astype(float)
movie.info()
```

```python
import numpy as np

movie_unique = pd.pivot_table(movie, index=['name'], aggfunc=np.sum) #영화별 평점 합산
movie_best = movie_unique.sort_values(by='point', ascending=False) # 합산된 평점을 내림차순 정렬
movie_best.head()
```
